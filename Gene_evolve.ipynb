```python
import entrezpy.esearch.esearcher
import urllib.request
import re
import sys
import os
os.system('clear')

```

```python
def esearch (query):
    
    e = entrezpy.esearch.esearcher.Esearcher('esearcher', 'onyekao@iastate.edu')
    a = e.inquire({'db':'protein','term':query, 'retmax': 100, 'retmode':'text','rettype': 'fasta'})

    id = a.get_result().uids[0]
    contents = urllib.request.urlopen("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=protein&id=%s&rettype=gp&retmode=text"%id).read().decode('utf-8')

    product = re.search('product="([\w\- ]+)"', contents).group(1)
    region = re.search('region_name="([\w\- ]+)"', contents).group(1)
    return ('%s "%s" "%s"'%(query,product,region))
    


    
```

```python
df['result']= df['Locuse'].apply(esearch)
```
